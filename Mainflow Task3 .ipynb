```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("householdtask3.csv")
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
#scatter plot with year against own
plt.scatter(df["year"], df["own"])

#adding title to the plot
plt.title("Scatter Plot")

#Setting the X and Y labels
plt.xlabel("year")
plt.ylabel("own")

#show the result
plt.show()
```

# Line chart

```python
#line chart with year against own 
plt.plot(df["year"])
plt.plot(df["own"])

#adding title to the chart
plt.title("Line Chart")

plt.xlabel("year")
plt.ylabel("own")

plt.show()
```

# Bar Chart

```python
#bar chart 
plt.bar(df["year"], df["income"])

#adding title to the chart
plt.title("Bar Chart")

#setting the x and y labels
plt.xlabel("year")
plt.ylabel("income")

plt.show()
```

```python
#using seaborn
```

```python
sns.barplot(x='year', y='income', data=df, color='green')  # Single color
plt.title('Bar Chart with Single Color')
plt.show()
```

```python
#Histogram
plt.hist(df["income"],bins=10,edgecolor='black')

plt.title("Histogram")


plt.show()
```
