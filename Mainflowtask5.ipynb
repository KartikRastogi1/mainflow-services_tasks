```python
#6 What is the most common cholesterol level (chol) in the dataset?
```

```python
# Find the most common cholesterol level (chol)
most_common_chol = df['chol'].mode()[0]

# Plotting the distribution of 'chol'
plt.figure(figsize=(8, 6))
sns.histplot(df['chol'], kde=True, color='lightgreen', bins=10)

# Highlight the most common cholesterol level

plt.axvline(most_common_chol, color='red', linestyle='-', linewidth=1)
plt.text(most_common_chol + 1, 1, f'Mode: {most_common_chol}', color='red')

# Add titles and labels
plt.title('Distribution of Cholesterol Levels (chol)')
plt.xlabel('Cholesterol Level (chol)')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
#7 What is the average resting blood pressure (trestbps) among the individuals?
average_trestbps = df['trestbps'].mean()

# Plotting the distribution of 'trestbps'
plt.figure(figsize=(10, 6))
sns.histplot(df['trestbps'], kde=True, color='lightcoral', bins=10)

# Highlight the average resting blood pressure
plt.axvline(average_trestbps, color='blue', linestyle='--', linewidth=2, label=f'Average: {average_trestbps:.2f}')
plt.text(average_trestbps + 1, 1, f'Average: {average_trestbps:.2f}', color='blue')

# Add titles and labels
plt.title('Distribution of Resting Blood Pressure (trestbps)')
plt.xlabel('Resting Blood Pressure (trestbps)')
plt.ylabel('Frequency')
plt.legend()

# Show the plot
plt.show()
```
